```python
import os
import cv2
import numpy as np
import random
from tensorflow.keras.utils import to_categorical

# Paths to real and fake videos
real_videos_path = "/content/drive/MyDrive/SDFVD2.0 Extension of Small Scale Deep Fake Video Dataset/SDFVD2.0 Extension of Small Scale Deep Fake Video Dataset/SDFVD2.0_real"
fake_videos_path = "/content/drive/MyDrive/SDFVD2.0 Extension of Small Scale Deep Fake Video Dataset/SDFVD2.0 Extension of Small Scale Deep Fake Video Dataset/SDFVD2.0_fake"

# Function to extract frames from video
def extract_frames(video_path, label, num_frames=5):
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Skip empty videos
    if total_frames == 0:
        print(f"‚ö†Ô∏è Skipping empty video: {video_path}")
        return []

    frames = []
    step = max(1, total_frames // num_frames)  # Ensure step is at least 1
    for i in range(0, total_frames, step):
        # Ensure frame_num is valid
        cap.set(cv2.CAP_PROP_POS_FRAMES, i)
        ret, frame = cap.read()

        if ret:
            frame = cv2.resize(frame, (128, 128))  # Resize for model
            frame = frame / 255.0  # Normalize
            frames.append((frame, label))
        else:
            print(f"‚ö†Ô∏è Frame not found at position {i} in {video_path}, skipping this frame.")

    cap.release()
    return frames

# Load dataset
data = []
video_count = 0

# Process real videos
for video in os.listdir(real_videos_path):
    if video.endswith(".mp4"):
        video_count += 1
        print(f"üì• Processing {video_count}: {video}")
        data.extend(extract_frames(os.path.join(real_videos_path, video), label=0))

# Process fake videos
for video in os.listdir(fake_videos_path):
    if video.endswith(".mp4"):
        video_count += 1
        print(f"üì• Processing {video_count}: {video}")
        data.extend(extract_frames(os.path.join(fake_videos_path, video), label=1))

# Check if dataset is empty
if not data:
    raise ValueError("‚ùå No video frames found. Check dataset paths and contents.")

# Shuffle data
random.shuffle(data)

# Convert to numpy arrays
X = np.array([item[0] for item in data])
y = np.array([item[1] for item in data])

# Convert labels to categorical (one-hot encoding)
y = to_categorical(y, num_classes=2)

# Split dataset into training (80%) and testing (20%)
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

print(f"‚úÖ Dataset loaded. Training Samples: {len(X_train)}, Testing Samples: {len(X_test)}")

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Build the CNN model
model = Sequential([
    Input(shape=(128, 128, 3)),  # Explicit Input layer
    Conv2D(32, (3,3), activation='relu', padding='same'),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu', padding='same'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(2, activation='softmax')  # Output: 2 classes (Real/Fake)
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("‚úÖ Model built successfully!")

# Train the model
model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))

# Save the model
model.save("deepfake_detector.h5")
print("‚úÖ Model trained and saved.")
```

```python
import gradio as gr
import cv2
import numpy as np
import tensorflow as tf

# Load the trained deepfake detection model
model = tf.keras.models.load_model("deepfake_detector.h5")

# Function to extract frames and make predictions
def predict_deepfake(video_path):
    cap = cv2.VideoCapture(video_path)
    frames = []

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_interval = max(1, total_frames // 10)  # Extract 10 frames evenly

    for i in range(10):
        cap.set(cv2.CAP_PROP_POS_FRAMES, i * frame_interval)
        ret, frame = cap.read()

        if ret:
            frame = cv2.resize(frame, (128, 128))  # Resize to model input size
            frame = frame / 255.0  # Normalize
            frames.append(frame)

    cap.release()

    if not frames:
        return "Error: No valid frames extracted"

    frames = np.array(frames)
    predictions = model.predict(frames)
    avg_prediction = np.mean(predictions, axis=0)  # Average prediction over frames

    result = "Real" if avg_prediction[0] > avg_prediction[1] else "Fake"
    return f"Prediction: {result}"

# Create Gradio Interface
iface = gr.Interface(
    fn=predict_deepfake,
    inputs=gr.Video(label="Upload a Video"),
    outputs=gr.Textbox(label="Deepfake Detection Result"),
    title="Deepfake Video Detector",
    description="Upload a video, and the AI will analyze it to determine if it's Real or Fake."
)

# Launch the Gradio app
iface.launch()
```
